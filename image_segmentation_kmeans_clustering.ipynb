# Image Segmentation

##  Download the necessary libraries

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

##  Download image and read the image

```python
img=plt.imread('2.jpeg')/255
```

## Check image shape

```python
img.shape
```

## Show the downloaded image

```python
plt.imshow(img)
plt.show()
```

## Convert the 3d image to 2d image

```python
im=img.reshape(img.shape[0]*img.shape[1],img.shape[2])
im.shape
```

# K-Means Modelling

## Choose the correct value of k using Elbow Curve

## Define the k-means function and fit the data

```python
l=[]
for i in range(1,11):
    km=KMeans(n_clusters=i,init='k-means++',random_state=0)
    km.fit(im)
    l.append(km.inertia_)
plt.plot(range(1,11),l,marker='o')
plt.title('elbow curve')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
kmeans.fit(im)
im_=kmeans.cluster_centers_[kmeans.labels_]
```

## Now Again plot the 2d image into 3d image

```python
clu_img=im_.reshape(img.shape[0],img.shape[1],img.shape[2])
plt.imshow(clu_img)
plt.show()
```
